```python
obj,num=[],[]
for col in df.columns:
  if df[col].dtype==object:
    obj.append(col)
  else:
    num.append(col)
obj,num
```

```python
for i in num:
  print(f'statistical analysis of column {i}')
  print(f'mean- {df[i].mean()}')
  print(f'median- {df[i].median()}')
  print(f'std- {df[i].std()}')
  print(f'min- {df[i].min()}')
  print(f'max- {df[i].max()}')
  print('--------------------------------')
```

```python
q1=df['Salary'].quantile(.25)
q3=df['Salary'].quantile(.75)
iqr=q3-q1
min_range=q1-iqr*1.5
maxrange=q3+iqr*1.5
```

```python
df=df[(df['Salary']>min_range)&(df['Salary']<maxrange)]
df.reset_index(inplace=True)
```

```python
def categorize_salary(salary):
    if salary < 400000:
        return 'Low'
    elif salary <= 800000:
        return 'Medium'
    else:
        return 'High'
```

```python
df['Salary Range'] = df['Salary'].apply(categorize_salary)
```

```python
def location_category(location):
    metro_cities = ['Bangalore', 'Mumbai', 'Delhi', 'Hyderabad', 'Chennai', 'Pune', 'Kolkata']

    if location in metro_cities :
      return 'Metropolitan'

    else:
      return 'Non-Metropolitan'
```

```python
df['Location Category'] = df['Location'].apply(location_category)
```
